# Analyse stock price with Moving Average

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Munging stock data from csv files

```python
ibm = pd.read_csv('../fin ta/data/ibm.csv')
ibm.head()
```

## Calculate and plot MA30 and MA200

```python
ibm['ma30'] = ibm['Adj Close'].rolling(30).mean()
ibm['ma200'] = ibm['Adj Close'].rolling(200).mean()

#plot the moving average
plt.figure(figsize=(10, 8))
ibm['ma30'].plot(label='MA30')
ibm['ma200'].plot(label='MA200')
ibm['Adj Close'].plot(label='Adj Close')
plt.legend()
plt.show()
```
