# Correlation between stocks

```python
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import stock prices

ibm = pd.read_csv('../fin ta/data/ibm.csv')
now = pd.read_csv('../fin ta/data/now.csv')
shop = pd.read_csv('../fin ta/data/shop.csv')
vmw = pd.read_csv('../fin ta/data/vmw.csv')
ibm.head()
adj_close = pd.DataFrame({'ibm':ibm['Adj Close'], 'now':now['Adj Close'], 'shop':shop['Adj Close'], 'vmw':vmw['Adj Close']})
adj_close.head()
```

```python
# Calculation of correlation each to each stock
adj_close.corr()
```

```python
# scatter matrix plot
sm = scatter_matrix(adj_close, figsize=(8, 8))
```

```python
# A closer look at NOW vs VMW.
adj_close.plot(kind='scatter', x='now', y='vmw', figsize=(8, 8))
```
